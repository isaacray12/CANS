# run

```python
# move into subject folder 
os.chdir(dataDir+sub_num)
# find csv file
wav_name = glob.glob('*.csv')[0][:-4]
# Get wave_dir directory from csv file
wave_dir = wav_name+'_mic_recorded'
# cd into directory
os.chdir(wave_dir)

# create_dic for storage of transcriptions
trans = {}

for wav in os.listdir():
    img_name = wav[10:22]+'.jpg'
    print(wav)
    with sr.AudioFile(wav) as source:
        # listen for the data (load audio to memory)
        audio_data = r.record(source)
        # recognize (convert from speech to text)
        text = r.recognize_google(audio_data)
        #x = r.recognize(audio,True) 
        print(img_name)
    trans[img_name] = text
    
```
